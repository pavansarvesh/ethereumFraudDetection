```python
from pathlib import Path
import pickle
import networkx as nx


def load_pickle(fname):
    with open(fname, 'rb') as f:
        return pickle.load(f)


pickle_path = Path('./dataset/ethereum-phishing-transaction-network/MulDiGraph.pkl')
if not pickle_path.exists():
    matches = list(Path('./dataset').rglob('MulDiGraph.pkl'))
    pickle_path = matches[0] if matches else None

if pickle_path is None:
    print('MulDiGraph.pkl not found under ./dataset')
    G = None
else:
    G = load_pickle(pickle_path)
    print(f'Graph with {G.number_of_nodes()} nodes and {G.number_of_edges()} edges')

if G is not None:
    # Traversal nodes:
    for idx, nd in enumerate(nx.nodes(G)):
        print(nd)
        print(G.nodes[nd]['isp'])
        break

    # Traversal edges:
    for ind, edge in enumerate(nx.edges(G)):
        (u, v) = edge
        eg = G[u][v][0]
        amo, tim = eg['amount'], eg['timestamp']
        print(amo, tim)
        break
```
